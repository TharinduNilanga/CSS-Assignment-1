```
def handle_exception(func):
  def wrapper():
    try:
      return func()
    except Exception as error:
      print("Invalid Operation",error)
      return None
  return wrapper

@handle_exception
def devide(a,b):
  return a/b

devide(10/0)
```

```
def handle_exception(func):
  def wrapper(*args,**kargs):
    try:
      return func(*args,**kwargs)
    except Exception as error:
      print("invalid Operation ",error)
      return None
    
  return wrapper

@handle_exception
def devide(a,b):
  return a/b
@handle_exception
def devide_c(a,b,c):
  return (a/b)/c
print(devide(10,4))
print(devide(10,0))

devide_c(10,2,0)
```

```

class Rectraingle:
  def __init__(self,length,width):
    self.length=length
    self.width=width
  def area(self):
    return self.length * self.width
rec = Rectraingle(10,4)
print("Rectangle Area ",rec.area())
```

```
class Square(Rectraingle):

  def __init__(self,length):
    super().__init__(length,length)

square = Square(100)
print("Square Area ",square.area())
```

```
class cube(Square):
  def surface_area(self):
    area=self.area()
    return area* 6

cu=cube(10)
print("cube surf area",cu.surface_area())
```
